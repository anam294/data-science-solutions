```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style(style="darkgrid")
```

```python
father_son = pd.read_csv('../data/father_son.csv').drop('Unnamed: 0', axis=1)
```

```python
father_son.head()
```

```python
y = (father_son['sheight'] - father_son['sheight'].mean()) / father_son['sheight'].std()
x = (father_son['fheight'] - father_son['fheight'].mean()) / father_son['fheight'].std()
```

```python
rho = np.corrcoef(x,y)[0][1]
rho
```

```python
'''
    draw the scatter plot
    axvline and axhline are used to draw the lines
'''
plt.figure(figsize=(10,8))
plt.scatter(x=x,y=y, c='salmon', alpha=0.2, s=12)
plt.xlim(-4, 4)
plt.ylim(-4, 4)
plt.plot([-4, 4], [-4, 4], 'k-', color='black', linewidth=1)
plt.axvline(x=0, color='black', linestyle='-', linewidth=1)
plt.axhline(y=0, color='black', linestyle='-', linewidth=1)
plt.plot(np.arange(-4, 4, 0.02), rho * np.arange(-4, 4, 0.02), 'k-', color='red', linewidth=2)
plt.plot(np.arange(-4, 4, 0.02), (1 / rho) * np.arange(-4, 4, 0.02), 'k-', color='red', linewidth=2)
```
